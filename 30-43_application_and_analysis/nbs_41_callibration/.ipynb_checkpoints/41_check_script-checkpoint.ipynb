```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xarray as xr
from datetime import datetime
from datetime import timedelta
import matplotlib.cm as cm
```

```python
path = Path.cwd()
figure_path = path / 'Figures'
```

```python
files = list((path / "Output" / "large").glob("*.nc"))
```

```python
files
```

```python
ds = xr.open_dataset(files[0])
```

```python
def calc_NSE(Qo, Qm):
    QoAv  = np.mean(Qo)
    ErrUp = np.sum((Qo - Qm)**2)
    ErrDo = np.sum((Qo - QoAv)**2)
    return 1 - (ErrUp / ErrDo)
    
def calc_log_NSE(Qo, Qm):
    Qo[Qo == 0] = 1e-6
    QoAv  = np.mean(Qo)
    ErrUp = np.sum((np.log(Qo) - np.log(Qm))**2)
    ErrDo = np.sum((np.log(Qo) - np.log(QoAv))**2)
    return 1 - (ErrUp / ErrDo)
```

```python
i = 0
NSE = np.zeros(ds.attrs['n_particles'])
log_NSE = np.zeros(ds.attrs['n_particles'])
for i in range(ds.attrs['n_particles']):
    NSE[i] = calc_NSE(ds["Q_obs"].values,ds["Q"].isel(EnsembleMember=i).values)
    log_NSE[i] = calc_log_NSE(ds["Q_obs"].values,ds["Q"].isel(EnsembleMember=i).values)
```

```python
plt.hist(NSE);
```

```python
i_max = NSE.argmax()
i_max_log  = log_NSE.argmax()
NSE.max()
```

```python
ds.isel(EnsembleMember=[i_max,i_max_log])
```

```python
log_NSE.max()
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,5))
for i in range(500):
    if i == i_max:
        ax.plot(ds["Q"].time.values, ds["Q"].sel(EnsembleMember=i).values,color="C1", zorder=10)    
    else:
        ax.plot(ds["Q"].time.values, ds["Q"].sel(EnsembleMember=i).values,color="C0")
    ds["Q_obs"].plot(ax=ax,color='k',lw=0,marker='.')
```

```python
df = pd.DataFrame(columns=["NSE_max","log_NSE_max","path"],)
for file in files:
    ds = xr.open_dataset(file)
    uid = ds.attrs['HRU_id']
    df.loc[uid, 'NSE_max'] = ds.attrs['NSE_max']
    df.loc[uid, 'log_NSE_max'] = ds.attrs['log_NSE_max']
    df.loc[uid,'path'] = file
    ds.close()
```

```python
fig, ax = plt.subplots(1,1)
ax.hist(df['NSE_max'],color="C0",alpha=0.5,label="NSE_max");
ax.hist(df['log_NSE_max'],color="C1",alpha=0.5, label="log_NSE_max");
ax.legend()
ax.set_ylabel('Number of catchments')
ax.set_xlabel("(log) NSE value");
title = f"Results callibrating 500 times for {len(df)} catchments"
ax.set_title(title);
title = title.replace(" ", '_')
fig.savefig(figure_path / (title + 'png'))
```
