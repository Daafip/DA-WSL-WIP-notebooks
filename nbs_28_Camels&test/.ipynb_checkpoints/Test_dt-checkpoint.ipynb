```python
import xarray as xr
import ewatercycle
import ewatercycle.util
from ewatercycle.util import get_time
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
path = Path.cwd()
forcing_path = path / "Forcing"
```

```python
experiment_start_date = "1981-01-01T00:00:00Z"
experiment_end_date = "1981-02-01T00:00:00Z"
```

```python
def crop_ds(ds: xr.Dataset, start_time: str, end_time: str) -> xr.Dataset:
    """Crops dataset based on time."""
    start = pd.Timestamp(get_time(experiment_start_date)).tz_convert(None)
    end = pd.Timestamp(get_time(end_time)).tz_convert(None)
    return ds.isel(
        time=(ds["time"].to_numpy() >= start) & (ds["time"].to_numpy() <= end)
    )
```

```python
ds = xr.open_dataset(forcing_path / 'HBV_forcing_CAMELS_2024-05-03 10_08.nc')
```

```python
ds.time.values[0]
```

```python
crop_ds(ds,experiment_start_date, experiment_end_date)
```
