```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = Path.cwd()
forcing_path = path / "Forcing"
observations_path = path / "Observations"
output_path = path / "Output"
paths = forcing_path, output_path, observations_path
figure_path = path / "Figures"
```

```python
def calc_NSE(Qo, Qm):
    QoAv  = np.mean(Qo)
    ErrUp = np.sum((Qm - Qo)**2)
    ErrDo = np.sum((Qo - QoAv)**2)
    return 1 - (ErrUp / ErrDo)
    
def calc_log_NSE(Qo, Qm):
    QoAv  = np.mean(Qo)
    ErrUp = np.sum((np.log(Qm) - np.log(Qo))**2)
    ErrDo = np.sum((np.log(Qo) - np.log(QoAv))**2)
    return 1 - (ErrUp / ErrDo)
```

```python
for i in range(len(files_df)):
    file = files_df[i]
    df_summary = pd.read_feather(file)
    NSE_mean = calc_NSE(df_summary['obs'].values, df_summary['mean'].values)
    NSE_mean_log = calc_log_NSE(df_summary['obs'].values, df_summary['mean'].values)
    fig, ax = plt.subplots(1,1,figsize=(12,5))
    ax.plot(df_summary['obs'],lw=0,marker="*",ms=2.0,zorder=0,label="Observations",color='k')
    ax.plot(df_summary['mean'],color="C1",lw=0.5,label=f"NSE μ {NSE_mean:.2f}\nNSE Log μ {NSE_mean_log:.2f}",zorder=-1)
    ax.fill_between(df_summary.index,df_summary['min'],df_summary['max'],color="C0", alpha=0.35,zorder=-10,label="bounds")
    ax.legend()
    params = file.name.split("_")[:6]
    params.pop(2)
    ax.set_title(f'HRU_id:{params[0]}, {params[1]}, {params[2]}, {params[3]},{params[4]}')
```
